```python
import pandas as pd 
df = pd.read_csv("2.csv") 
df.head() 
df.shape 
```

```python

# computes various summary statistics, excluding NaN values 
df.describe() 
# for computing correlations 
df.corr() 
# computes numerical data ranks 
df.rank()
```

```python
import matplotlib.pyplot as plt 
# plot a histogram  
df['gender'].hist(bins=10) 
  
# shows presence of a lot of outliers/extreme values 
df.boxplot(column='gender', by = 'ms') 
  
# plotting points as a scatter plot 
x = df["gender"] 
y = df["ms"] 
plt.scatter(x, y, label= "stars", color= "m",  
            marker= "*", s=30) 
# x-axis label 
plt.xlabel('gender') 
# frequency label 
plt.ylabel('ms') 
# function to show the plot 
plt.show() 
```
